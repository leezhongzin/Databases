Arbitrary Shaped Circles

```python
from sklearn import cluster, datasets, mixture
from sklearn.datasets import make_circles
from sklearn import metrics
```

```python
#Creating and Visualising dataset
n_samples = 1500
random_state = 170
circles = datasets.make_circles(n_samples=n_samples, random_state=8)
X, y = datasets.make_circles(n_samples=n_samples, factor=0.5, noise=0.05)

agg_clust = AgglomerativeClustering(n_clusters=3)
assigned_clust = agg_clust.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=assigned_clust, s=10);
```

```python
agg_comp = AgglomerativeClustering(linkage ='complete', n_clusters=3)
agg_avg = AgglomerativeClustering(linkage ='average', n_clusters=3)
agg_ward = AgglomerativeClustering(linkage ='ward', n_clusters=3)
as_comp = agg_comp.fit_predict(X)
as_avg = agg_avg.fit_predict(X)
as_ward = agg_ward.fit_predict(X)
```

Visualizing the predictions made by the model using different linkage methods

```python
plt.scatter(X[:, 0], X[:, 1], c=as_comp, s=10);
```

```python
plt.scatter(X[:, 0], X[:, 1], c=as_avg, s=10)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = as_ward, s = 10)
```
